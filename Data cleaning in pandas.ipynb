```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\hp\Downloads\Customer Call List.xlsx")
df
```

# Removing Dublicates 

```python
df = df.drop_duplicates()
df
```

# Dropping Columns

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

# Strip

```python
#df["Last_Name"] = df["Last_Name"].str.strip("/")
#df["Last_Name"] = df["Last_Name"].str.strip("_")
#df["Last_Name"] = df["Last_Name"].str.strip("...")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

# Cleaning/Standardizing phone_number

```python
#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[4:7] + '-' + x[8:12])
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[4:7] + '-' + x[8:12])
df["Phone_Number"] = df ["Phone_Number"].replace('nan---','')
df["Phone_Number"] = df["Phone_Number"].replace('NaN','')
df["Phone_Number"] = df["Phone_Number"].replace('N/a','')
df["Phone_Number"] = df["Phone_Number"].replace('N/a---','')
df["Phone_Number"] = df["Phone_Number"].replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].replace('--','')
df
```

# Splitting Columns

```python
df[["Street Address","State","Zip_Code"]] = df["Address"].str.split(',' , n=2 ,expand = True)
df
```

# Standarizing column value using replace

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yeseseseses','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('N/a','N')
df["Paying Customer"] = df["Paying Customer"].str.replace('Y','Yes')
df["Paying Customer"] = df["Paying Customer"].str.replace('N','No')
df
```

# Fill Null Values

```python
#df = df.replace('Yes','Y')
df = df.replace('No','N')
df = df.replace('NaN','')
df = df.fillna('')
df
```

# Filtering Down Of Rows

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == '':
        df.drop(x,inplace=True)
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x,inplace=True)
        
# another way to drop null values        
#df = df.dropna(subset ="Phone_Number"), inplace = True)        
```
