# importing libraries and  methods

```python
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

# inputting the URL

```python
url = 'https://wuzzuf.net/search/jobs/?a=navbg%7Cspbg&q=illustrator'
```

# create a client based request to get the url

```python
client = urlopen(url)
```

# getting the HTML code off the full page

```python
html = client.read()
```

```python
html
```

# closing the request

```python
client.close()
```

# create an HTML parser using beautifulsoup

```python
soup = bs(html,"html.parser")
```

```python
soup
```

# creating a container for the needed data

```python
containers = soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})
```

```python
len(containers)
```

```python
bs.prettify(containers[0])
```

# bringing it all together

```python
f = open("wuzzuf-illustrator.csv", "w")
header = "jop_title, company_name, jop_location, jop_type\n"
f.write(header)
```

```python
for container in containers:
    jtitle =  container.find_all("h2",{"class","css-m604qf"})
    jop_title = jtitle[0].text.strip()
    
    cname = container.findAll("a",{"class":"css-17s97q8"})
    company_name = cname[0].text.strip()

    jtype = container.findAll("span",{"class":"css-1ve4b75 eoyjyou0"})
    jop_type = jtype[0].text.strip()

#    print(jop_title)
#    print(company_name)
#    print(jop_type)
#    print()

#    print(jop_title + ", " + company_name + ", "  + ", " + jop_type)
#    print()

f.write(jop_title + ", " + company_name + ", "  + ", " + jop_type + "\n") 
f.close()
```
